# 2.4 Manipulando Dados

```python
# Adicionando colunas
df['Ano_Analise'] = 2023
```

```python
# Adicionando coluna com condição
import numpy as np
```

```python
df['Cobrança_Abusiva']= np.where( # Where() funciona como if-else
    df["charges"] > 20000,    
    "Sim",       #retorno para True
    "Não"        #retorno para False

)
df.head()
```

```python
# Exclusão de coluna
# Não alteram o objeto efetivamente, mas nos dão um view de como eles são com aquela modificação.
df.drop(columns= "Ano_Analise")
```

```python
df.head()
```

```python
# Para realmente alterar o objeto é necessário utilizar o parâmetro inplace das funções.
df.drop(columns= "Ano_Analise", inplace=True)
df.head()
```

```python
#Troca Sim por 1 e Não por 0
df['Cobrança_Abusiva'] = df['Cobrança_Abusiva'].map({
    "Sim":1,
    "Não":0
})

#Cria coluna com valor 2023
df['Ano_Analise'] = 2023

#Atualiza valor 2023 para 2022
df['Ano_Analise'] = 2022

df.head()
```

# 2.5 Analisando Dados

```python
#Estatística descritiva
df.describe()
```

```python
#count: Mostra o número de observações (linhas) não-nulas em cada coluna do DataFrame.
#mean: Calcula a média aritmética dos valores em cada coluna. É a soma de todos os valores dividida pelo número de valores.
#std: Representa o desvio padrão, que mede a quantidade de variação ou dispersão de um conjunto de valores. Um desvio padrão baixo indica que os valores tendem a estar próximos da média; um desvio padrão alto indica que os valores estão espalhados por uma gama mais ampla de valores.
#min: Mostra o valor mínimo em cada coluna do DataFrame.
#25%: Representa o primeiro quartil (Q1) dos dados, ou seja, 25% dos dados são menores que esse valor. É calculado ordenando os dados e tomando o valor no índice que é 25% do caminho através da lista.
#50%: Representa a mediana ou o segundo quartil (Q2) dos dados. Metade dos dados são menores que esse valor. É o valor no meio de um conjunto de dados ordenado.
#75%: Representa o terceiro quartil (Q3) dos dados, ou seja, 75% dos dados são menores que esse valor. É o valor abaixo do qual se encontram 75% dos dados.
#max: Mostra o valor máximo em cada coluna do DataFrame.
```

```python
# O describe() serve apenas para colunas númericas.  Se quisermos olhar as categóricas, vamos utilizar value_counts().
df["sex"].value_counts() #quantidade de pessoas de cada sexo
df["sex"].value_counts(normalize=True) #Percentual
```

```python
#Cruzando as variáveis
df.groupby("sex")["age"].mean()   #grupamos a idade por sexo e depois tiramos a média usando a função mean()
```

```python
# Média de idade e cobrança entre fumantes e não fumantes dos custos médicos
df.groupby("smoker") [["age","charges"]].mean()
```

```python
# Desvio Padrão por sexo e fumante
df.groupby(["smoker", "sex"]) ["charges"].std()
```
