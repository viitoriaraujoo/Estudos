# Adicionando Detalhes ao Grfico

```python
# Cria uma contagem
smoke_counts= insurance['smoker'].value_counts()

# Tamanho do Grfico
plt.figure(figsize=(8,6))

# Cria grfico de barras com indice e contagem
plt.bar(
    smoke_counts.index, # Valor do eixo x
    smoke_counts.values, # Valor do eixo y
    color= ['steelblue', 'lightcoral'] # Cores das barras
)

# Rtulo do Eixo X com fonte de tamanho 8
plt.xlabel('Smoker', fontsize=8)

# Rtulo do Eixo Y com fonte de tamanho 8
plt.ylabel('Count', fontsize=8)

# Ttulo do Grfico com fonte de tamanho 14
plt.title('Distribuio de Fumantes', fontsize=14)

# Remove as linhas do topo e da direita
plt.gca().spines['top'].set_visible(False) #get current axes
plt.gca().spines['right'].set_visible(False) 

plt.show()
```

# Uso do subplot na criao de mltiplos grficos

```python
# Criao do local, configurao do plot

fig,axes = plt.subplots(
    1, #Nmero de linhas
    2, #Nmero de colunas
    figsize=(12, 6) #Tamanho do Plot
)

# Grfico de Disperso de BMI x Charges na posio 0 
# (Linha 0, Coluna 0)
axes[0].scatter(insurance['bmi'], insurance['charges'])
axes[0].set_xlabel('BMI') # Define o rtulo do eixo x do primeiro subplot
axes[0].set_ylabel('charges') #Define o rtulo do eixo y do primeiro subplot
axes[0].set_title('Charges vs BMI')

# Grfico de Disperso de BMI x Charges na posio 1 
# (Linha 0, Coluna 1)
axes[1].scatter(insurance['age'], insurance['charges'])
axes[1].set_xlabel('age')
axes[1].set_ylabel('charges')
axes[1].set_title('Charges vs Age')

# Ajuste de espao entre os plots
plt.tight_layout # Ajusta automaticamente o espaamento entre subplots para que eles no se sobreponham.

# Apresenta o grfico
plt.show()
```

# Anlise envolvendo varivel categrica e uma numrica

```python

# Calcular as mdias das cobranas para cada categoria de 'sex'
mean_charges_sex = insurance.groupby('sex')['charges'].mean()

# Calcular as mdias das cobranas para cada categoria de 'region'
mean_charges_region = insurance.groupby('region')['charges'].mean()

# Calcular as mdias das cobranas para cada categoria de 'children'
mean_charges_children = insurance.groupby('children')['charges'].mean()

# Calcular as mdias das cobranas para cada categoria de 'smoker'
mean_charges_smoker = insurance.groupby('smoker')['charges'].mean()

# Cores Personalizadas para as barras
colors = ['#2b8cbe', '#fe9929', '#31a354', '#a1d99b']

# Criar uma figura com subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Grfico de barras: mdia das cobranas por sexo
mean_charges_sex.plot(kind='bar', ax=axes[0, 0], color=colors)
axes[0, 0].set_xlabel('Sexo')
axes[0, 0].set_ylabel('Cobranas Mdias')
axes[0, 0].set_title('Cobranas Mdias por Sexo')

# Grfico de barras: mdia das cobranas por regio
mean_charges_region.plot(kind='bar', ax=axes[0, 1], color=colors)
axes[0, 1].set_xlabel('Regio')
axes[0, 1].set_ylabel('Cobranas Mdias')
axes[0, 1].set_title('Cobranas Mdias por Regio')

# Grfico de barras: mdia das cobranas por filhos
mean_charges_children.plot(kind='bar', ax=axes[1, 1], color=colors)
axes[1, 1].set_xlabel('Nmero de Filhos')
axes[1, 1].set_ylabel('Cobranas Mdias')
axes[1, 1].set_title('Cobranas Mdias por Nmero de Filhos')

# Grfico de barras: mdia das cobranas por fumante
mean_charges_smoker.plot(kind='bar', ax=axes[1, 0], color=colors)
axes[1, 0].set_xlabel('Fumante')
axes[1, 0].set_ylabel('Cobranas Mdias')
axes[1, 0].set_title('Cobranas Mdias por Fumante')

# Ajustar espaamento entre os subplot
plt.tight_layout()

# Exibir o Grfico
plt.show()
```

# Alterando Propriedades com 'rcParams'

```python
# Carrega a biblioteca
import matplotlib as mpl

# Definindo o tamanho padro dos grficos
mpl.rcParams['figure.figsize'] =[10,5]

# Definindo a espessura das linhas
mpl.rcParams['lines.linewidth'] =2

# Definindo o tamanho da fonte dos ttulos dos eixos
mpl.rcParams['axes.titlesize'] =14

# Definindo o tamanho da fonte dos rtulos dos eixos
mpl.rcParams['axes.labelsize'] =12

# Definindo o tamanho da fonte dos rtulos dos ticks
mpl.rcParams['xtick.labelsize'] =12

# Definindo o tamanho da fonte da legenda
mpl.rcParams['legend.fontsize']=12

# Definindo a resoluo dos grficos
mpl.rcParams['figure.dpi'] = 150

#Para printar o dicionrio rcParams

for key in mpl.rcParams.keys():
    print(key)
```

```python
import matplotlib.pyplot as plt #Biblioteca para visualizao de grficos

# Cria uma contagem para cada tipo de regio
region_types = insurance['region'].value_counts() # Contagem de cada valor nico

# Listamos as regies existentes
region_name = ['southwest', 'southeast', 'northwest', 'northeast']

# Personaliza as cores para cada barra
colors = ['steelblue', 'lightcoral', 'mediumseagreen', 'orange']

# Cria uma Figura e um eixo com largura de 8 polegadas e 5 de altura
fig, ax = plt.subplots(figsize=(8, 6)) 

# Plota o grfico de barras com as cores personalizadas
ax.bar(region_name, region_types.values, color=colors)

# Configurao dos rtulos dos eixos e ttulos
ax.set_xlabel('Regio')
ax.set_ylabel('Contagem')
ax.set_title('Contagem de Regies')

# Adicionar a contagem sobre cada barra
for i, v in enumerate(region_types.values): #posio no eixo x e y
    ax.text(i, v, str(v), ha='center', va='bottom') 

# O loop facilita o raciocinio, pois seno, teriamos que realizar manualmente com ax.text
    
    
# Exibe o grfico
plt.show()
```

```python
# Calcular a mdia da cobrana por sexo
mean_charges_sex = insurance.groupby('sex')['charges'].mean()

# Personalizao das cores para cada barra
colors = ['blue', 'pink']

# Cria uma Figura e um Eixo
fig, ax = plt.subplots(figsize=(8, 6))  # Corrigi o erro aqui, adicionando o sinal de igual

# Plota o grfico de barras com as cores personalizadas
ax.bar(mean_charges_sex.index, mean_charges_sex.values, color=colors)

# Configurao dos rtulos dos eixos e ttulo
ax.set_xlabel('Sexo')  # Corrigi o nome do mtodo para definir o rtulo do eixo x
ax.set_ylabel('Cobranas')  # Corrigi o nome do mtodo para definir o rtulo do eixo y
ax.set_title('Mdia das Cobranas por Sexo')

# Adiciona manualmente os textos dos rtulos
for i, value in enumerate(mean_charges_sex.values):
    ax.text(i, value, f'{value:.2f}', ha='center', va='bottom', fontsize=12)
    
# Define limites dos eixos x e y
ax.set_xlim(-0.5, 1.5)
ax.set_ylim(0, 16000)

# Calcula a diferena percentual entre os dois sexos
diff_percentual = ((mean_charges_sex['male'] - mean_charges_sex['female']) / mean_charges_sex['female']) * 100

# Mensagem sobre a diferena salarial
ax.text(
    1.2, #posio no eixo x
    mean_charges_sex.values[1]+1000, #posio no eixo y
    f'0 salrio dos homens  {diff_percentual:.2f}% maior' #mensagem
)

# Exibe o grfico
plt.show()
```
