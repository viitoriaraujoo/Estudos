# Questão 2 - O seu chefe decidiu criar duas categorias, Premium e Gold, para focar nos clientes mais promissores. Para a categoria Premium, entrarão os que possuem saldo (balance) nos top 10%. Ou seja, o grupo acima do percentil 90. Já o grupo Gold, ficarão os que possuem saldo nos top 20%, mas abaixo do grupo Premium. Ou seja, do p80 até o último do p90. Verifique quantos clientes possuem no grupo Premium e no Gold.

```python
# Calcular os percentis
p80 = stead_original['balance'].quantile(0.8)
p90 = stead_original['balance'].quantile(0.9)
```

```python
# Filtrar clientes para a categoria Premium (acima do percentil 90)
premium_clientes = stead_original[stead_original['balance'] > p90]
print(f'Número de clientes na categoria Premium: {len(premium_clientes)}')
```

```python
# Filtrar clientes para a categoria Gold (entre o percentil 80 e o último do percentil 90)
gold_clientes = stead_original[(stead_original['balance'] >= p80) & (stead_original['balance'] <= p90)]
print(f'Número de clientes na categoria Gold: {len(gold_clientes)}')
```

# Questão 3 - Para a apresentação final, você decide criar uma tabela com os 10 clientes que possuem maior saldo (balance). Como a tabela é grande demais, você resolve apresentar apenas as informações relevantes: age, job, marital, education, default, loan, duration. Além disso, você decide destacar clientes que foram mais e menos atenciosos, pois são os que parecem ter maior e menor potencial, respectivamente. Para isso, destaque na tabela, de amarelo, o cliente de maior duration. Destaque, de vermelho, o que tem menor duration.



```python
import pandas as pd
colunas_selecionadas = ['age', 'job', 'marital', 'education', 'default', 'loan', 'duration', 'balance']
stead = stead_original[colunas_selecionadas]

# Obter as 10 linhas com os maiores valores na coluna 'balance'
top_10_clientes = stead.nlargest(10, 'balance')

# Define a função que destaca máximo e mínimo
def destacar_linha(s):
    is_max = s.index == top_10_clientes['duration'].idxmax()
    is_min = s.index == top_10_clientes['duration'].idxmin()
    return ['background-color: yellow' if v_max else 'background-color: red' if v_min else '' for v_max, v_min in zip(is_max, is_min)]

# Aplica formatação condicional
styled_top_10_clientes = top_10_clientes.style.apply(destacar_linha)

# Exibir a tabela destacada
styled_top_10_clientes
```

## Questão 5 - Você trabalha numa empresa de bicicletas e precisa fazer uma clusterização separando marcas caras de baratas. Para isso, você precisará saber a média de preços das bicicletas de cada marca.


```python
# Carregando o DATASET Products
arquivo = r'C:\Users\casa\Downloads\Cursos\Projeto Pandas\Bike Store Relational Database\products.csv'
products= pd.read_csv(arquivo, sep= ',') #indica o ponto e virgula como separador
products

# Carregando o brands Products
arquivo_2= r'C:\Users\casa\Downloads\Cursos\Projeto Pandas\Bike Store Relational Database\brands.csv'
brands= pd.read_csv(arquivo_2, sep= ',') #indica o ponto e virgula como separador
brands
```

```python
# Mesclar os DataFrames usando a coluna 'brand_id'
merged_df = pd.merge(products, brands, on='brand_id')

# Calcular a média de preços por marca
average_prices = merged_df.groupby('brand_name')['list_price'].mean().reset_index()
average_prices
```

## Questão 6 - Agora que você já possui o preço médio de cada marca, você deve separar os valores em 3 grupos, sendo que os nomes de cada agrupamento será a partir dessa lista: ["Barato", "Médio", "Caro"].


```python
# Definir os rótulos de faixa de preço
bins = [average_prices['list_price'].min(), average_prices['list_price'].quantile(1/3), average_prices['list_price'].quantile(2/3), average_prices['list_price'].max()]
labels = ["Barato", "Médio", "Caro"]

# Criar uma nova coluna 'faixa_preco' baseada nos rótulos
average_prices['faixa_preco'] = pd.cut(average_prices['list_price'], bins=bins, labels=labels, include_lowest=True)

# Exibir o DataFrame resultante
average_prices
```
